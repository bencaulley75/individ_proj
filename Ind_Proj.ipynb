```python
plt.figure()
plt.bar(df['name'], df['protein'], label='Protein')
plt.bar(df['name'], df['total_fat'], bottom=df['protein'], label='Total Fat')
plt.bar(df['name'], df['energy'], bottom=df['protein'] + df['total_fat'], label='Energy')
plt.xlabel('Food Item')
plt.ylabel('Content')
plt.title('Macronutrient Composition')
plt.xticks(rotation=90)
_ = plt.legend()

```
